```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal=np.random.normal(loc=0,scale=1,size=1000);
for i in range(101):
    print(i,np.percentile(std_normal,i));
```

```python
msmnts1=np.random.normal(loc=0,scale=1,size=100);
stats.probplot(msmnts1,dist="norm",plot=pylab);
pylab.show();
```

```python
#There are deviation of points from line. This is because of less number of observations.
#We increase the observations, and then see.
```

```python
msmnts2=np.random.normal(loc=0,scale=1,size=1000);
stats.probplot(msmnts2,dist="norm",plot=pylab);
pylab.show();
```

```python
#Now we can see that the points lie on the line
#Checking for uniform random var and normal
```

```python
msmnts3=np.random.uniform(low=-1,high=1,size=1000);
stats.probplot(msmnts3,dist="norm",plot=pylab);
pylab.show();
```

```python
#The dist of msmnts3 is not normal as is shown by the figure.
```
