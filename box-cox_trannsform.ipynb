```python
from scipy import stats;
import matplotlib.pyplot as plt;
import pylab
```

```python
# fig=pylab.figure();
# ax1=fig.add_subplot(211);
```

```python
x=stats.loggamma.rvs(5,size=500)+5;
#print(x)
stats.probplot(x,dist=stats.norm, plot=pylab);
pylab.show();
```

```python
#Using boxcox transform
#ax2=fig.add_subplot(212);
xt,_=stats.boxcox(x);
stats.probplot(xt, dist=stats.norm, plot=pylab);
pylab.show()
```

```python
# x which was loggamma which follows power law or pareto distribution,  is converted to normal using boxcox.
```
